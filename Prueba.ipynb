Escribe un programa que calcule la resistencia equivalente a un circuito de tres resistencias en paralelo como el que se muestra en la siguiente figura:



```python
R1: float = float(input('Ingrese Valor Resistencia R1'))
R2: float = float(input('Ingrese Valor Resistencia R2'))
R3: float = float(input('Ingrese Valor Resistencia R3'))

Rserie = R1 + R2 + R3
RParalelo = (1/R1) + (1/R2) + (1/R3)

print(f'La Resistencia en Serie es {Rserie} Ohm y la Resistencia en Paralelo es {RParalelo:.2f} Ohm')
```
