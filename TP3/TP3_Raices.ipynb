### Mtodo Intervalo Medio
*Aplicado a* \( `f(x) = 24x^3 - 2x^2 - 5x + 1` \)


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funcin a partir de la entrada del usuario
fun = input("Introduzca la funcin f(x) (use 'x' como variable): ")
f = lambda x: eval(fun)

# Graficar la funcin
x_vals = np.linspace(-1, 1, 1000)
y_vals = f(x_vals)
plt.figure(figsize=(20,20))
plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfica de la funcin')
plt.grid(True)
plt.legend()
plt.show()

# Obtener los lmites
n = int(input('Cuntos intervalos desea usar? '))
intervalos = []
for i in range(n):
    a = float(input(f'Introduzca el valor de Lim Inf para el intervalo {i + 1}: '))
    b = float(input(f'Introduzca el valor de Lim Sup para el intervalo {i + 1}: '))
    intervalos.append((a, b))

IntAux = intervalos.copy()
tol = float(input('Introduzca la Tolerancia: '))

for idx, (a, b) in enumerate(IntAux):  # Usar IntAux aqu
    if a > b:
        print(f'Error: El intervalo [{a}, {b}] no es vlido. Asegrese de que Lim Inf < Lim Sup.')
        continue
    
    if f(a) * f(b) >= 0:
        print(f"Error: f(a) y f(b) deben tener signos opuestos para el intervalo [{a}, {b}].")
    else:
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        error_aux = 1
        while error_aux > tol:  # Cota de error
            c_ant = c
            if f(c) == 0:
                print(f'La raz exacta es: {c} en el intervalo [{IntAux[idx]}]')
                break
            elif f(a) * f(c) < 0:
                b = c  # La raz est en [a, c]
            else:
                a = c  # La raz est en [c, b]
            c = (a * f(b) - b * f(a)) / (f(b) - f(a))
            error_aux = abs(c - c_ant)
        
        print(f'La raz aproximada es: {c:.5f} en el intervalo [{IntAux[idx][0]}, {IntAux[idx][1]}]')
```

*Aplicado a* \( `f(x) = 0.0018x + 0.3x^2 - 0.0013x^3 - 372` \)

```python
fun = input("Introduzca la funcin f(x) (use 'x' como variable): ")
f = lambda x: eval(fun)

# Graficar la funcin
x_vals = np.linspace(0, 100, 1000)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfica de la funcin')
plt.grid(True)
plt.legend()
plt.show()

# Obtener los lmites inferior y superior
a = float(input('Introduzca el valor de Lim Inf: '))
b = float(input('Introduzca el valor de Lim Sup: '))
tol = float(input('Introduzca la Tolerancia: '))

# Mtodo de Intervalo Medio
if f(a) * f(b) >= 0:
    print("Error: f(a) y f(b) deben tener signos opuestos.")
else:
    max_iter = int(np.log((b - a) / tol) / np.log(2))
    for k in range(max_iter):
        c = (a + b) / 2
        if f(a) * f(c) < 0:
            b = c
        elif f(a) * f(c) == 0:
            print(f'La raz es: {c}')
            break
        else:
            a = c
    else:
        print(f'La raz aproximada es: {c}')
```

### Resultados del Mtodo de Intervalo Medio

Despus de aplicar el **Mtodo de Intervalo Medio** a la funcin:

\[ 
f(x) = 0.0018x + 0.3x^2 - 0.0013x^3 - 372 
\]

hemos encontrado que la raz aproximada, que representa la cantidad ptima de impresoras a producir y vender, es:

Esto sugiere que el fabricante debera considerar la produccin de aproximadamente **39 impresoras** al mes para maximizar su utilidad.
