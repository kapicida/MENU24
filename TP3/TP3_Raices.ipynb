### Ejercicio 3


* Metodo Intervalo Medio

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a partir de la entrada del usuario
fun = input("Introduzca la función f(x) (use 'x' como variable): ")
f = lambda x: eval(fun)

# Graficar la función
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función')
plt.grid(True)
plt.legend()
plt.show()

# Obtener los límites inferior y superior
a = float(input('Introduzca el valor de Lim Inf: '))
b = float(input('Introduzca el valor de Lim Sup: '))
tol = float(input('Introduzca la Tolerancia: '))

# Método de Intervalo Medio
if f(a) * f(b) >= 0:
    print("Error: f(a) y f(b) deben tener signos opuestos.")
else:
    max_iter = int(np.log((b - a) / tol) / np.log(2))
    for k in range(max_iter):
        c = (a + b) / 2
        if f(a) * f(c) < 0:
            b = c
        elif f(a) * f(c) == 0:
            print(f'La raíz es: {c}')
            break
        else:
            a = c
    else:
        print(f'La raíz aproximada es: {c}')
```

### Ejercicio 4



* Metodo Interpolacion Lineal

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a partir de la entrada del usuario
fun = input("Introduzca la función f(x) (use 'x' como variable): ")
f = lambda x: eval(fun)

# Graficar la función
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función')
plt.grid(True)
plt.legend()
plt.show()

a = float(input('Introduzca el valor de Lim Inf: '))
b = float(input('Introduzca el valor de Lim Sup: '))
tol = float(input('Introduzca la Tolerancia: '))

if f(a) * f(b) >= 0:
    print("Error: f(a) y f(b) deben tener signos opuestos.")
else:
    while (b - a) / 2 > tol:  # Cota de error
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))  # Interpolación lineal
        if f(c) == 0:
            print(f'La raíz exacta es: {c}')
            break
        elif f(a) * f(c) < 0:
            b = c  # La raíz está en [a, c]
        else:
            a = c  # La raíz está en [c, b]
    else:
        print(f'La raíz aproximada es: {c}')
```

```python
import numpy as nps
import matplotlib.pyplot as plt

def intervalo_medio(f, a, b, tol):
    if f(a) * f(b) >= 0:
        raise ValueError("Error: f(a) y f(b) deben tener signos opuestos")
    else:
        iteraciones = 0
        while (b - a) / 2.0 > tol:
            c = (a + b) / 2.0
            if f(c) == 0:
                return c, iteraciones, 0.0
            elif f(a) * f(c) < 0:
                b = c
            else:
                a = c
            iteraciones += 1
            error = (b - a) / 2.0  
        return (a + b) / 2.0, iteraciones, error

# Funcion 1
def f1(x):
    return 24 * x**3 - 2 * x**2 - 5 * x + 1
tol1 = 0.0001
a1, b1 = -2, 2 

#Funcion 2
def f2(x):
    return 0.0013 * (x**3) + 0.3 * (x**2) + 8 * x - 372
tol2 = 0.001
a2, b2 = 0, 100

raiz1, iteraciones1,error1 = intervalo_medio(f1, a1, b2, tol1)
print(f"F1: Raíz encontrada: {raiz1}\n(Error:{error1})\n(Error R:{error1/abs(raiz1)})\n(Iteraciones: {iteraciones1})")

raiz2, iteraciones2,error2 = intervalo_medio(f2, a2, b2, tol2)
print(f"F2: Raíz encontrada: {raiz2}\n(Error:{error2})\n(Error R:{error2/abs(raiz2)})\n"
     f"(Iteraciones: {iteraciones2})")
```
